# Tokenizers from hugging face

we can use tokenizer of any pretrained model from hugging face 

we have and import from tranformers import AutoTokenizer

then 

tokenizer = AutoTokenizer.from_pretrained(model)

simple 

to encode we use tokenizer.encode(string)
to decode we use same tokenizer.decode(token_id) or .batch_decode(token_ids)

to convert our regular message to passable prompt we use tokenizer.apply_template(message) which converts the regular list of dicts to prompt with generation token as per models requirements
