## Today’s Learnings

Today, I learned how to call a model by going deeper into the code without using pipelines. We explored the roles of the tokenizer, model, encoder, and streamer in more detail.

We also studied the structure of transformer models more closely. A model typically includes:

- An embedding layer
- Several decoder layers
- Within each decoder:
  - Query projection
  - Key projection
  - Value projection
  - Output projection

We also looked at the gating components used in the model, including:

- Gate projection
- Up projection
- Down projection
- Activation function

In addition, we discussed the importance of normalization layers, which help keep the model’s activations stable and balanced. We also learned about quantization and why non-linearity is so important. Without non-linearity, all the layered parameters would not be able to learn meaningful representations. I also learned that non-linearity is introduced through activation functions.
